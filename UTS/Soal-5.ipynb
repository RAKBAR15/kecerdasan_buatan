```python
import json

def get_company_employee_nationalities():
    nationality_counts_by_company = {}

    for employee in employee_detail_list:
        company = employee['company']
        nationality = employee['nationality']
        
        if company not in nationality_counts_by_company:
            nationality_counts_by_company[company] = {}
        
        if nationality in nationality_counts_by_company[company]:
            nationality_counts_by_company[company][nationality] += 1
        else:
            nationality_counts_by_company[company][nationality] = 1
    
    result = [
        {
            "company": company,
            "employee_nationality": nationality_counts
        }
        for company, nationality_counts in nationality_counts_by_company.items()
    ]
    
    return result
```

```python
output = get_company_employee_nationalities()
```
